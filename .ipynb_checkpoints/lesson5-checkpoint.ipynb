```python
# File path
points_fp = r"data/addresses_geo.shp"

# Read the data
points = gpd.read_file(points_fp)
```

```python
points
```

```python
def getPointCoords(row, geom, coord_type):
    """Calculates coordinates ('x' or 'y') of a Point geometry"""
    if coord_type == 'x':
        return row[geom].x
    elif coord_type == 'y':
        return row[geom].y
```

```python
# Calculate x coordinates
points['x'] = points.apply(getPointCoords, geom='geometry', coord_type='x', axis=1)

# Calculate y coordinates
points['y'] = points.apply(getPointCoords, geom='geometry', coord_type='y', axis=1)

# Let's see what we have now
points.head()
```

```python
p_df = points.drop('geometry', axis=1).copy()
p_df.head(2)
```

```python
from bokeh.models import ColumnDataSource

# Point DataSource
psource = ColumnDataSource(p_df)

# What is it?
psource
```

```python
# Initialize our plot figure
p = figure(title="A map of address points from a Shapefile")

# Add the points to the map from our 'psource' ColumnDataSource -object
p.circle('x', 'y', source=psource, color='red', size=10)
```

```python
# Output filepath
outfp = r"point_map.html"

# Save the map
save(p, outfp)
```

```python
from bokeh.models import HoverTool
my_hover = HoverTool()
```

```python
# ESEMPIO
# my_hover2.tooltips = [('Label1', '@col1'), ('Label2', '@col2'), ('Label3', '@col3')]
my_hover.tooltips = [('Address of the point', '@address')]
```

```python
p.add_tools(my_hover)
```

```python
# Output filepath
outfp = r"point_map_hover.html"

# Save the map
save(p, outfp)
```

```python
admin2_file = r'MozambiqueFlood/Mocambique_admin2.shp'

# Read the data
admin2 = gpd.read_file(admin2_file)
admin2.head()
```

```python
from bokeh.models import GeoJSONDataSource

# Convert GeoDataFrames into GeoJSONDataSource objects (similar to ColumnDataSource)
adm2_source = GeoJSONDataSource(geojson=admin2.to_json())

# Initialize our plot figure
p = figure(title="A test map")

# Add the lines to the map from our GeoJSONDataSource -object (it is important to specify the columns as 'xs' and 'ys')
p.patches('xs', 'ys', source=adm2_source, color='gray', line_color='white', line_width=2)

adm2_hover = HoverTool()
# my_hover2.tooltips = [('Label1', '@col1'), ('Label2', '@col2'), ('Label3', '@col3')]
adm2_hover.tooltips = [('Admin2 Name', '@Admin2')]

p.add_tools(adm2_hover)

# Output filepath
outfp = r"test_map_geojson.html"

# Save the map
save(p, outfp)
```
